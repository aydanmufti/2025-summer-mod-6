```python
# Homework Reflection 5: Causal Diagrams and Feedback Loops
# Clean version without emojis to avoid font warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.patches import FancyBboxPatch, ConnectionPatch
import warnings
warnings.filterwarnings('ignore')

# Set style for better plots
plt.style.use('default')
plt.rcParams['figure.figsize'] = [10, 6]
plt.rcParams['font.size'] = 10
np.random.seed(42)

print("Homework Reflection 5: Causal Diagrams and Feedback Loops")
print("=" * 70)

# %% [markdown]
## Problem 1: Negative Feedback Loop - Body Temperature and Sweating
# 
# **Situation:**
# - Sweating causes body temperature to decrease
# - High body temperature causes sweating
# 
# **Challenge:** We need a DAG (no cycles), but this describes a feedback loop!
# 
# **Solution:** Use time subscripts to break the cycle

# %%
def draw_feedback_diagram():
    """Draw the negative feedback loop using time subscripts"""
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 7))
    
    # Left plot: The problematic cycle (what we CAN'T draw in a DAG)
    ax1.set_xlim(0, 10)
    ax1.set_ylim(0, 10)
    ax1.set_title("INVALID: Cyclic Graph (Not Allowed in DAG)", fontsize=14, fontweight='bold', color='red')
    
    # Draw nodes
    temp_pos = (3, 7)
    sweat_pos = (7, 3)
    
    # Temperature node
    temp_box = FancyBboxPatch((temp_pos[0]-1.2, temp_pos[1]-0.6), 2.4, 1.2, 
                              boxstyle="round,pad=0.1", facecolor='lightcoral', edgecolor='black', linewidth=2)
    ax1.add_patch(temp_box)
    ax1.text(temp_pos[0], temp_pos[1], 'Body\nTemperature', ha='center', va='center', fontweight='bold', fontsize=11)
    
    # Sweating node
    sweat_box = FancyBboxPatch((sweat_pos[0]-1.2, sweat_pos[1]-0.6), 2.4, 1.2,
                               boxstyle="round,pad=0.1", facecolor='lightblue', edgecolor='black', linewidth=2)
    ax1.add_patch(sweat_box)
    ax1.text(sweat_pos[0], sweat_pos[1], 'Sweating', ha='center', va='center', fontweight='bold', fontsize=11)
    
    # Draw arrows (creating a cycle)
    ax1.annotate('', xy=(sweat_pos[0]-1.2, sweat_pos[1]+0.4), xytext=(temp_pos[0]+1.2, temp_pos[1]-0.4),
                arrowprops=dict(arrowstyle='->', lw=3, color='red'))
    ax1.text(5, 5.8, 'High temp\ncauses sweating', ha='center', fontweight='bold', color='red', fontsize=10)
    
    ax1.annotate('', xy=(temp_pos[0]+1.2, temp_pos[1]-0.8), xytext=(sweat_pos[0]-1.2, sweat_pos[1]+0.8),
                arrowprops=dict(arrowstyle='->', lw=3, color='blue'))
    ax1.text(5, 4.2, 'Sweating decreases\ntemperature', ha='center', fontweight='bold', color='blue', fontsize=10)
    
    ax1.set_xticks([])
    ax1.set_yticks([])
    ax1.text(5, 1, 'WARNING: This creates a cycle!\nNot allowed in DAG', ha='center', fontsize=12, 
             bbox=dict(boxstyle="round,pad=0.4", facecolor='yellow', alpha=0.8), fontweight='bold')
    
    # Right plot: The DAG solution using time subscripts
    ax2.set_xlim(0, 10)
    ax2.set_ylim(0, 10)
    ax2.set_title("VALID: DAG Solution Using Time Subscripts", fontsize=14, fontweight='bold', color='green')
    
    # Time t nodes
    temp_t_pos = (2, 7)
    sweat_t_pos = (2, 4)
    
    # Time t+1 nodes
    temp_t1_pos = (8, 7)
    sweat_t1_pos = (8, 4)
    
    # Draw nodes for time t
    temp_t_box = FancyBboxPatch((temp_t_pos[0]-1, temp_t_pos[1]-0.5), 2, 1,
                                boxstyle="round,pad=0.1", facecolor='lightcoral', edgecolor='black', linewidth=2)
    ax2.add_patch(temp_t_box)
    ax2.text(temp_t_pos[0], temp_t_pos[1], 'Temp(t)', ha='center', va='center', fontweight='bold', fontsize=11)
    
    sweat_t_box = FancyBboxPatch((sweat_t_pos[0]-1, sweat_t_pos[1]-0.5), 2, 1,
                                 boxstyle="round,pad=0.1", facecolor='lightblue', edgecolor='black', linewidth=2)
    ax2.add_patch(sweat_t_box)
    ax2.text(sweat_t_pos[0], sweat_t_pos[1], 'Sweat(t)', ha='center', va='center', fontweight='bold', fontsize=11)
    
    # Draw nodes for time t+1
    temp_t1_box = FancyBboxPatch((temp_t1_pos[0]-1, temp_t1_pos[1]-0.5), 2, 1,
                                 boxstyle="round,pad=0.1", facecolor='lightcoral', edgecolor='black', linewidth=2)
    ax2.add_patch(temp_t1_box)
    ax2.text(temp_t1_pos[0], temp_t1_pos[1], 'Temp(t+1)', ha='center', va='center', fontweight='bold', fontsize=11)
    
    sweat_t1_box = FancyBboxPatch((sweat_t1_pos[0]-1, sweat_t1_pos[1]-0.5), 2, 1,
                                  boxstyle="round,pad=0.1", facecolor='lightblue', edgecolor='black', linewidth=2)
    ax2.add_patch(sweat_t1_box)
    ax2.text(sweat_t1_pos[0], sweat_t1_pos[1], 'Sweat(t+1)', ha='center', va='center', fontweight='bold', fontsize=11)
    
    # Draw arrows for the DAG
    # Temp(t) -> Sweat(t+1)
    ax2.annotate('', xy=(sweat_t1_pos[0]-1, sweat_t1_pos[1]), xytext=(temp_t_pos[0]+1, temp_t_pos[1]),
                arrowprops=dict(arrowstyle='->', lw=3, color='red'))
    ax2.text(5, 6, 'High temp causes\nsweating later', ha='center', fontweight='bold', color='red', fontsize=10)
    
    # Sweat(t) -> Temp(t+1)
    ax2.annotate('', xy=(temp_t1_pos[0]-1, temp_t1_pos[1]), xytext=(sweat_t_pos[0]+1, sweat_t_pos[1]),
                arrowprops=dict(arrowstyle='->', lw=3, color='blue'))
    ax2.text(5, 5, 'Sweating decreases\ntemp later', ha='center', fontweight='bold', color='blue', fontsize=10)
    
    # Persistence arrows
    ax2.annotate('', xy=(temp_t1_pos[0]-1, temp_t1_pos[1]-0.6), xytext=(temp_t_pos[0]+1, temp_t_pos[1]-0.6),
                arrowprops=dict(arrowstyle='->', lw=2, color='gray', linestyle='--'))
    ax2.annotate('', xy=(sweat_t1_pos[0]-1, sweat_t1_pos[1]+0.6), xytext=(sweat_t_pos[0]+1, sweat_t_pos[1]+0.6),
                arrowprops=dict(arrowstyle='->', lw=2, color='gray', linestyle='--'))
    ax2.text(5, 8.5, 'Persistence over time', ha='center', fontsize=9, color='gray', style='italic')
    
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax2.text(5, 1.5, 'SUCCESS: Now it is a valid DAG!\nNo cycles present', ha='center', fontsize=12,
             bbox=dict(boxstyle="round,pad=0.4", facecolor='lightgreen', alpha=0.8), fontweight='bold')
    
    plt.tight_layout()
    plt.show()

draw_feedback_diagram()

print("\nProblem 1 Solution:")
print("To represent negative feedback in a DAG, we use time subscripts:")
print(" Temp(t)  Sweat(t+1): High temperature causes sweating")
print(" Sweat(t)  Temp(t+1): Sweating decreases temperature")
print(" This breaks the cycle and maintains the DAG property!")

# %% [markdown]
## Problem 2: Positive Feedback Loop Example

print("\n" + "="*70)
print("Problem 2: Positive Feedback Loop Example")
print("="*70)

def draw_positive_feedback():
    """Draw an example of positive feedback loop"""
    fig, ax = plt.subplots(1, 1, figsize=(14, 8))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_title("Positive Feedback Loop: Social Media Engagement", fontsize=16, fontweight='bold')
    
    # Time t nodes
    posts_t_pos = (2, 7)
    followers_t_pos = (2, 3)
    
    # Time t+1 nodes  
    posts_t1_pos = (8, 7)
    followers_t1_pos = (8, 3)
    
    # Draw nodes
    nodes = [
        (posts_t_pos, 'Posts(t)', 'lightblue'),
        (followers_t_pos, 'Followers(t)', 'lightgreen'),
        (posts_t1_pos, 'Posts(t+1)', 'lightblue'),
        (followers_t1_pos, 'Followers(t+1)', 'lightgreen')
    ]
    
    for pos, label, color in nodes:
        box = FancyBboxPatch((pos[0]-1, pos[1]-0.5), 2, 1,
                             boxstyle="round,pad=0.1", facecolor=color, edgecolor='black', linewidth=2)
        ax.add_patch(box)
        ax.text(pos[0], pos[1], label, ha='center', va='center', fontweight='bold', fontsize=11)
    
    # Draw arrows
    # More posts -> more followers
    ax.annotate('', xy=(followers_t1_pos[0]-1, followers_t1_pos[1]), 
                xytext=(posts_t_pos[0]+1, posts_t_pos[1]),
                arrowprops=dict(arrowstyle='->', lw=4, color='red'))
    ax.text(5, 5.8, 'More posts \nMore followers', ha='center', fontweight='bold', color='red', fontsize=11)
    
    # More followers -> more posts
    ax.annotate('', xy=(posts_t1_pos[0]-1, posts_t1_pos[1]), 
                xytext=(followers_t_pos[0]+1, followers_t_pos[1]),
                arrowprops=dict(arrowstyle='->', lw=4, color='blue'))
    ax.text(5, 4.2, 'More followers \nMore motivation to post', ha='center', fontweight='bold', color='blue', fontsize=11)
    
    # Persistence arrows
    ax.annotate('', xy=(posts_t1_pos[0]-1, posts_t1_pos[1]-0.6), 
                xytext=(posts_t_pos[0]+1, posts_t_pos[1]-0.6),
                arrowprops=dict(arrowstyle='->', lw=2, color='gray', linestyle='--'))
    ax.annotate('', xy=(followers_t1_pos[0]-1, followers_t1_pos[1]+0.6), 
                xytext=(followers_t_pos[0]+1, followers_t_pos[1]+0.6),
                arrowprops=dict(arrowstyle='->', lw=2, color='gray', linestyle='--'))
    
    ax.set_xticks([])
    ax.set_yticks([])
    
    # Add explanation
    ax.text(5, 1, 'POSITIVE FEEDBACK: Both variables reinforce each other upward!\n' +
                  'More posts  More followers  More motivation  Even more posts', 
            ha='center', fontsize=12, fontweight='bold',
            bbox=dict(boxstyle="round,pad=0.5", facecolor='yellow', alpha=0.8))
    
    plt.tight_layout()
    plt.show()

draw_positive_feedback()

print("Positive Feedback Example: Social Media Engagement")
print(" More posts  More followers (content attracts audience)")
print(" More followers  More posts (motivation from larger audience)")
print(" Result: Exponential growth in both variables!")
print()
print("Other examples of positive feedback:")
print(" Economic bubbles: Higher prices  More buying  Even higher prices")
print(" Population growth: More individuals  More reproduction  Even more individuals")
print(" Viral content: More shares  More visibility  Even more shares")

# %% [markdown]
## Problem 3: Ecosystem Diagram with Lightning, Deer, Bears, and Flowers

print("\n" + "="*70)
print("Problem 3: Ecosystem Causal Diagram")
print("="*70)

def draw_ecosystem_diagram():
    """Draw the ecosystem diagram"""
    fig, ax = plt.subplots(1, 1, figsize=(14, 10))
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_title("Ecosystem Causal Diagram", fontsize=16, fontweight='bold')
    
    # Node positions
    lightning_pos = (5, 8.5)
    deer_pos = (2, 5)
    bears_pos = (8, 5)
    flowers_pos = (5, 2)
    
    # Draw nodes
    nodes = [
        (lightning_pos, 'Lightning\nStorms', 'yellow'),
        (deer_pos, 'Deer\nPopulation', 'lightgreen'),
        (bears_pos, 'Bear\nPopulation', 'sandybrown'),
        (flowers_pos, 'Flower\nPopulation', 'lightpink')
    ]
    
    for pos, label, color in nodes:
        box = FancyBboxPatch((pos[0]-0.9, pos[1]-0.7), 1.8, 1.4,
                             boxstyle="round,pad=0.15", facecolor=color, edgecolor='black', linewidth=2)
        ax.add_patch(box)
        ax.text(pos[0], pos[1], label, ha='center', va='center', fontweight='bold', fontsize=12)
    
    # Draw arrows with labels
    arrows = [
        # Lightning effects
        (lightning_pos, deer_pos, 'red', 'Frighten away\n(DECREASE)', '(-)'),
        (lightning_pos, bears_pos, 'red', 'Frighten away\n(DECREASE)', '(-)'),
        (lightning_pos, flowers_pos, 'green', 'Promote growth\n(INCREASE)', '(+)'),
        
        # Predation
        (bears_pos, deer_pos, 'darkred', 'Eat deer\n(DECREASE)', '(-)'),
        (deer_pos, flowers_pos, 'orange', 'Eat flowers\n(DECREASE)', '(-)')
    ]
    
    for start, end, color, label, sign in arrows:
        # Calculate arrow position
        dx = end[0] - start[0]
        dy = end[1] - start[1]
        length = np.sqrt(dx**2 + dy**2)
        dx_norm, dy_norm = dx/length, dy/length
        
        start_adj = (start[0] + 0.9*dx_norm, start[1] + 0.7*dy_norm)
        end_adj = (end[0] - 0.9*dx_norm, end[1] - 0.7*dy_norm)
        
        ax.annotate('', xy=end_adj, xytext=start_adj,
                   arrowprops=dict(arrowstyle='->', lw=4, color=color))
        
        # Add label at midpoint
        mid_x, mid_y = (start[0] + end[0])/2, (start[1] + end[1])/2
        ax.text(mid_x, mid_y, f'{sign}\n{label}', ha='center', va='center', 
               fontweight='bold', fontsize=10, color=color,
               bbox=dict(boxstyle="round,pad=0.3", facecolor='white', alpha=0.9, edgecolor=color))
    
    ax.set_xticks([])
    ax.set_yticks([])
    
    # Add explanation
    explanation = """Causal Relationships:
Lightning  Deer/Bears (negative): Storms frighten animals away
Lightning  Flowers (positive): Storms help flowers grow  
Bears  Deer (negative): Bears eat deer
Deer  Flowers (negative): Deer eat flowers"""
    
    ax.text(5, 0.5, explanation, ha='center', va='center', fontsize=11, fontweight='bold',
           bbox=dict(boxstyle="round,pad=0.4", facecolor='lightblue', alpha=0.8))
    
    plt.tight_layout()
    plt.show()

draw_ecosystem_diagram()

# Now let's simulate this ecosystem
def simulate_ecosystem_data(n=1000, seed=42):
    """Simulate the ecosystem data with the specified relationships"""
    np.random.seed(seed)
    
    # Lightning storms (frequency per year)
    lightning = np.random.poisson(3, n)  # Average 3 storms per year
    
    # Baseline populations (without lightning effects)
    deer_baseline = np.random.normal(50, 10, n)  # Average 50 deer
    bear_baseline = np.random.normal(20, 5, n)   # Average 20 bears  
    flower_baseline = np.random.normal(100, 20, n) # Average 100 flower units
    
    # Lightning effects
    # Lightning frightens away deer and bears (negative effect)
    deer_pop = np.maximum(1, deer_baseline - lightning * 5 + np.random.normal(0, 5, n))
    bear_pop = np.maximum(1, bear_baseline - lightning * 2 + np.random.normal(0, 3, n))
    
    # Lightning promotes flower growth (positive effect)  
    flower_pop = np.maximum(1, flower_baseline + lightning * 10 + np.random.normal(0, 8, n))
    
    # Bears eat deer (negative effect of bears on deer)
    deer_pop = np.maximum(1, deer_pop - bear_pop * 0.3 + np.random.normal(0, 3, n))
    
    # Deer eat flowers (negative effect of deer on flowers)
    flower_pop = np.maximum(1, flower_pop - deer_pop * 0.8 + np.random.normal(0, 5, n))
    
    return pd.DataFrame({
        'lightning_storms': lightning,
        'deer_population': deer_pop,
        'bear_population': bear_pop, 
        'flower_population': flower_pop
    })

# Generate the ecosystem data
print("Generating ecosystem simulation data...")
ecosystem_df = simulate_ecosystem_data(1000, seed=42)

print("\nEcosystem Data Summary:")
print(ecosystem_df.describe().round(2))

# Show the data generation code
print("\nData Generation Code Structure:")
print("""
def simulate_ecosystem_data(n=1000, seed=42):
    # Lightning storms (Poisson distribution, avg 3 per year)
    lightning = np.random.poisson(3, n)
    
    # Baseline populations (normal distributions)
    deer_baseline = np.random.normal(50, 10, n)
    bear_baseline = np.random.normal(20, 5, n)
    flower_baseline = np.random.normal(100, 20, n)
    
    # Lightning effects
    deer_pop = max(1, deer_baseline - lightning*5 + noise)     # Lightning scares deer
    bear_pop = max(1, bear_baseline - lightning*2 + noise)     # Lightning scares bears
    flower_pop = max(1, flower_baseline + lightning*10 + noise) # Lightning helps flowers
    
    # Predation effects
    deer_pop = max(1, deer_pop - bear_pop*0.3 + noise)        # Bears eat deer
    flower_pop = max(1, flower_pop - deer_pop*0.8 + noise)    # Deer eat flowers
    
    # All effects include random noise for realism
""")

# Identify backdoor paths
print("\nBackdoor Path Analysis:")
print("="*50)
print("RESEARCH QUESTION: What is the causal effect of DEER on FLOWERS?")
print()
print("Direct path: Deer  Flowers (deer eat flowers)")
print()
print("Backdoor paths (confounding paths):")
print("1. Deer  Lightning  Flowers")
print("   - Lightning affects both deer (negatively) and flowers (positively)")
print("   - This creates a confounding relationship")
print()
print("2. Deer  Lightning  Bears  Deer (already affects deer directly)")
print() 
print("CONFOUNDERS IDENTIFIED: Lightning storms")
print("To estimate the causal effect of deer on flowers, we must control for lightning!")

# Visualize the relationships
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Lightning vs Deer
axes[0,0].scatter(ecosystem_df['lightning_storms'], ecosystem_df['deer_population'], alpha=0.6, color='green')
axes[0,0].set_xlabel('Lightning Storms per Year')
axes[0,0].set_ylabel('Deer Population')
axes[0,0].set_title('Lightning  Deer (Negative Effect)')
axes[0,0].grid(True, alpha=0.3)

# Lightning vs Flowers  
axes[0,1].scatter(ecosystem_df['lightning_storms'], ecosystem_df['flower_population'], alpha=0.6, color='pink')
axes[0,1].set_xlabel('Lightning Storms per Year')
axes[0,1].set_ylabel('Flower Population')
axes[0,1].set_title('Lightning  Flowers (Positive Effect)')
axes[0,1].grid(True, alpha=0.3)

# Deer vs Flowers (confounded!)
axes[1,0].scatter(ecosystem_df['deer_population'], ecosystem_df['flower_population'], alpha=0.6, color='orange')
axes[1,0].set_xlabel('Deer Population')
axes[1,0].set_ylabel('Flower Population')
axes[1,0].set_title('Deer  Flowers (CONFOUNDED by Lightning!)')
axes[1,0].grid(True, alpha=0.3)

# Bears vs Deer
axes[1,1].scatter(ecosystem_df['bear_population'], ecosystem_df['deer_population'], alpha=0.6, color='brown')
axes[1,1].set_xlabel('Bear Population')
axes[1,1].set_ylabel('Deer Population')
axes[1,1].set_title('Bears  Deer (Negative Effect)')
axes[1,1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# %% [markdown]
## Problem 4: Custom Causal Diagram with Confounder and Collider

print("\n" + "="*70)
print("Problem 4: Custom Causal Diagram")
print("="*70)

def draw_custom_diagram():
    """Draw a custom causal diagram with confounder and collider"""
    fig, ax = plt.subplots(1, 1, figsize=(16, 11))
    ax.set_xlim(0, 12)
    ax.set_ylim(0, 10)
    ax.set_title("Custom Causal Diagram: Education and Career Success", fontsize=18, fontweight='bold')
    
    # Node positions
    education_pos = (3, 5)      # Treatment (X)
    income_pos = (9, 5)         # Outcome (Y)
    family_wealth_pos = (6, 8)  # Confounder
    job_satisfaction_pos = (6, 2)  # Collider
    networking_pos = (6, 5)     # Mediator
    
    # Draw nodes
    nodes = [
        (education_pos, 'Education\nLevel', 'lightblue'),
        (income_pos, 'Annual\nIncome', 'lightgreen'), 
        (family_wealth_pos, 'Family\nWealth', 'gold'),
        (job_satisfaction_pos, 'Job\nSatisfaction', 'lightcoral'),
        (networking_pos, 'Professional\nNetworking', 'lightpink')
    ]
    
    for pos, label, color in nodes:
        # Make treatment and outcome nodes larger
        if 'Education' in label or 'Income' in label:
            width, height = 2.4, 1.6
            fontsize = 13
            linewidth = 3
        else:
            width, height = 2.2, 1.4
            fontsize = 11
            linewidth = 2
            
        box = FancyBboxPatch((pos[0]-width/2, pos[1]-height/2), width, height,
                             boxstyle="round,pad=0.15", facecolor=color, edgecolor='black', linewidth=linewidth)
        ax.add_patch(box)
        ax.text(pos[0], pos[1], label, ha='center', va='center', fontweight='bold', fontsize=fontsize)
    
    # Draw arrows
    arrows = [
        # Confounding path: Family Wealth affects both Education and Income
        (family_wealth_pos, education_pos, 'blue', 'Enables better\neducation', 3.0),
        (family_wealth_pos, income_pos, 'blue', 'Direct wealth\neffects', 3.0),
        
        # Treatment to mediator to outcome
        (education_pos, networking_pos, 'green', 'Better education\nmore connections', 2.5),
        (networking_pos, income_pos, 'green', 'Connections\nbetter jobs', 2.5),
        
        # Direct effect  
        (education_pos, income_pos, 'red', 'Skills\nhigher pay', 4.0),
        
        # Collider: both Education and Income affect Job Satisfaction
        (education_pos, job_satisfaction_pos, 'purple', 'Education\nfulfillment', 2.5),
        (income_pos, job_satisfaction_pos, 'purple', 'Money\nhappiness', 2.5),
    ]
    
    for start, end, color, label, width in arrows:
        # Calculate arrow direction
        dx = end[0] - start[0] 
        dy = end[1] - start[1]
        length = np.sqrt(dx**2 + dy**2)
        
        # Adjust start and end points to node boundaries
        if length > 0:
            dx_norm, dy_norm = dx/length, dy/length
            start_adj = (start[0] + 1.2*dx_norm, start[1] + 0.8*dy_norm)
            end_adj = (end[0] - 1.2*dx_norm, end[1] - 0.8*dy_norm)
            
            ax.annotate('', xy=end_adj, xytext=start_adj,
                       arrowprops=dict(arrowstyle='->', lw=width, color=color))
            
            # Add label
            mid_x = (start[0] + end[0]) / 2
            mid_y = (start[1] + end[1]) / 2
            
            # Offset label to avoid overlap
            if abs(dx) > abs(dy):  # Horizontal arrow
                offset_y = 0.5 if dy >= 0 else -0.5
                offset_x = 0
            else:  # Vertical arrow
                offset_x = 1.0 if dx >= 0 else -1.0
                offset_y = 0
                
            ax.text(mid_x + offset_x, mid_y + offset_y, label, ha='center', va='center',
                   fontweight='bold', fontsize=10, color=color,
                   bbox=dict(boxstyle="round,pad=0.25", facecolor='white', alpha=0.95, edgecolor=color))
    
    # Add legend/explanation
    legend_text = """TREATMENT (X): Education Level
OUTCOME (Y): Annual Income

CONFOUNDER: Family Wealth
(affects both education and income)

COLLIDER: Job Satisfaction  
(affected by both education and income)
      
MEDIATOR: Professional Networking
(education  networking  income)"""
    
    ax.text(0.5, 5, legend_text, ha='left', va='center', fontsize=12, fontweight='bold',
           bbox=dict(boxstyle="round,pad=0.5", facecolor='lightyellow', alpha=0.95, edgecolor='black'))
    
    ax.set_xticks([])
    ax.set_yticks([])
    
    plt.tight_layout()
    plt.show()

draw_custom_diagram()

print("Custom Scenario: Education and Career Success")
print()
print("Variables:")
print(" Education Level (X) - TREATMENT")
print(" Annual Income (Y) - OUTCOME") 
print(" Family Wealth - CONFOUNDER")
```
